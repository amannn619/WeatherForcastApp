```python
import requests
from datetime import datetime
```

```python
lat = 2.183333 
lon = 33.683333
data = requests.get(f"http://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid=7b02f9f1a037bf6848cc8f4240c4989c")
```

```python
data = data.json()
```

```python
data["list"]
```

```python
for i in data["list"]:
    temp = i["main"]["temp"]
    weather = i["weather"][0]["main"]
    date =  datetime.strptime(i["dt_txt"], "%Y-%m-%d %H:%M:%S")    
    print(date, weather, temp)
    
```
